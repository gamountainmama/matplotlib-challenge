## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

drug_mean = clean_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
drug_median = clean_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
drug_var = clean_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
drug_std = clean_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
drug_sem = clean_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()

drug_summary = pd.DataFrame({
    "Mean Tumor Volume": drug_mean,
    "Median Tumor Volume": drug_median,
    "Tumor Volume Variance": drug_var,
    "Tumor Volume Std. Dev.": drug_std,
    "Tumor Volume Std. Err.": drug_sem})
drug_summary
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.
clean_data.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":[("mean", "mean"),("median", "median"),
                                                              ("var", "var"),("std","std"),("sem","sem")]})
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_time = clean_data.groupby("Mouse ID")["Timepoint"].max()
last_time

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
tumor_last = pd.merge(last_time, clean_data, how = "left", on = ["Mouse ID","Timepoint"])
tumor_last.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for x in treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    vol_data.append(tumor_last["Tumor Volume (mm3)"].loc[tumor_last["Drug Regimen"] == x])
    
    # add subset 
    quartiles = vol_data[0].quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    
    # Determine outliers using upper and lower bounds
    lower = lowerq - 1.5 * iqr
    upper = upperq + 1.5 * iqr
    
    print(f"For {x}, values below {round(lower,2)} or above {round(upper,2)} could be outliers.")
    outliers = []
    
    for y in vol_data[0]:
        if y > upper or y < lower:
            outliers.append(y)
            print(f"For {x}, {round(outliers[0],2)} is an outlier.")
            print("")
    
    if outliers == []:
        print(f"For {x}, there are no outliers.")
        print("")
    
    vol_data = []
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
capomulin = tumor_last.loc[tumor_last["Drug Regimen"] == "Capomulin"]
ramicane = tumor_last.loc[tumor_last["Drug Regimen"] == "Ramicane"]
infubinol = tumor_last.loc[tumor_last["Drug Regimen"] == "Infubinol"]
ceftamin = tumor_last.loc[tumor_last["Drug Regimen"] == "Ceftamin"]

tumor_summary = pd.DataFrame({
    "Capomulin": capomulin["Tumor Volume (mm3)"],
    "Ramicane": ramicane["Tumor Volume (mm3)"],
    "Infubinol": infubinol["Tumor Volume (mm3)"],
    "Ceftamin": ceftamin["Tumor Volume (mm3)"]})
tumor_summary

tumor_summary.boxplot(grid = False, flierprops={'marker': '*', 'markersize': 20, 'markerfacecolor': 'fuchsia'})
plt.ylabel("Final Tumor Volume (mm3)")
plt.title("Boxplots of Final Tumor Volume (mm3) for Each Drug Regimen")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
weight = capomulin_avg["Weight (g)"]
avg_tumor = capomulin_avg["Tumor Volume (mm3)"]
correlation = st.pearsonr(weight, avg_tumor)
print(f"The correlation between mouse weight and average tumor volume is {round(correlation[0],2)}")

(slope, intercept, rvalue, pvalue, stderr) = linregress(weight,avg_tumor)
regress = weight * slope + intercept
capomulin_avg.plot(kind = "scatter", x = "Weight (g)", y = "Tumor Volume (mm3)", xlabel = "Weight (g)",
                  ylabel = "Average Tumor Volume (mm3)", title = "Average Tumor Volume (mm3) by Weight")
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.annotate(line_eq,(20,37),fontsize=15,color="red")
plt.plot(weight, regress, color = "red")
plt.show()
```
